# JitCoder / Codify Project Checklist

Concise, file-by-file action plan to make the project safer, cleaner, and more production-ready.


## 1) Repo hygiene

**Files:** `.gitignore`, root repo

- Remove `.env*`, `db.sqlite3`, `node_modules/`, `dist/`, `staticfiles/`, virtual env folders
- Add `.env.example` files for backend and frontend
- Save `requirements.txt` as UTF-8


## 2) Secure test cases

**Files:** `contest/models.py`, `contest/serializers.py`, `contest/views.py`

- Add `is_sample` field to `TestCase`
- Expose only sample test cases to normal users
- Keep hidden/private test cases for judge only
- Use separate public/private serializers if needed


## 3) Restrict write permissions

**Files:** `contest/views.py`

- Make `Problem`, `Contest`, `ContestProblem`, `TestCase` write access admin-only
- Keep student submission endpoints authenticated-only
- Add custom permission like `IsAdminOrReadOnly`


## 4) Fix auth serializer issues

**Files:** `account/serializers.py`, `account/models.py`

- Remove wrong validation imports
- Use `serializers.ValidationError`
- Ensure serializers match custom user fields (`email`, `name`, not `username` unless present)
- Prefer `AbstractBaseUser + PermissionsMixin`


## 5) Split run vs submit flow

**Files:** `judge/views.py`, `judge/services.py`, `judge/urls.py`

- `run-code/` = custom input or sample run
- `submit-code/` = judge against all test cases
- Move judge helper logic into `judge/services.py`
- Keep views thin


## 6) Correct leaderboard logic

**Files:** `judge/tasks.py`

- Do not rank by latest submission only
- Use best submission per user per problem
- Rebuild leaderboard from accepted/best results
- Keep ordering deterministic


## 7) Fix WebSocket contract

**Files:** `contest/consumers.py`, websocket broadcast code, `frontend/src/hooks/useContestSocket.js`

- Standardize message shape with `event` and `data`
- Align backend event names with frontend listeners
- Add handlers for leaderboard, submission, participant count


## 8) Unify frontend API layer

**Files:** `frontend/my-app/src/lib/api.js`, `src/api/*.js`, `.env`

- Create one shared Axios client
- Use `VITE_API_BASE_URL` and `VITE_WS_BASE_URL`
- Remove hardcoded localhost URLs
- Avoid double `/api/api/...` paths


## 9) Persist contest join status

**Files:** `contest/models.py`, `contest/views.py`, `contest/serializers.py`

- Add `ContestParticipation` model
- Save joined state in DB
- Return `joined` in contest serializer
- Use DB state, not only local frontend state


## 10) Timezone + Redis cleanup

**Files:** `codify/settings.py`

- Use consistent timezone (`Asia/Kolkata`)
- Keep `USE_TZ=True`
- Move Redis URL to env var
- Reuse same Redis URL for Celery and Channels


## 11) Password reset + frontend URL

**Files:** `account/views.py`

- Build reset links using `FRONTEND_URL` env var
- Remove hardcoded localhost URLs


## 12) Build + deploy safety

**Files:** `build.sh`, deployment config

- `pip install -r requirements.txt`
- `python manage.py migrate`
- `python manage.py collectstatic --noinput`
- `python manage.py check --deploy`


## 13) Theme consistency

**Files:** `frontend/my-app/src/styles/theme.css`, page CSS

- Replace hardcoded light-only colors
- Use CSS variables for cards, text, borders, tables, tags
- Make dark mode consistent across dashboard, problem pages, tags, tables


## 14) UX polish

**Files:** `App.jsx`, pages/components

- Add 404 page
- Add loading + error states
- Show submission history
- Improve editor usability and responsiveness


## 15) Suggested implementation order

### Critical first
1. Repo cleanup
2. Hide private test cases
3. Restrict permissions
4. Unify API base URLs
5. Fix websocket payloads

### Then
6. Split run / submit flow
7. Fix leaderboard scoring
8. Persist joined state
9. Redis + timezone cleanup
10. Theme + UX polish
